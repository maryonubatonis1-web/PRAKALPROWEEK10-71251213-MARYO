```python
def tiga_terbaik(data):
    data_urut = sorted(data, reverse=True)
    return data_urut[:3]
angka = [10, 50, 80, 200, 30, 150]
print(tiga_terbaik(angka))
```

```python
total = 0
count = 0

while True:
    data = input("Masukkan bilangan (ketik 'done' untuk selesai): ")
    
    if data.lower() == "done":
        break
    
    try:
        angka = float(data)
        total += angka
        count += 1
    except:
        print("Input tidak valid, masukkan angka!")

if count > 0:
    rata_rata = total / count
    print("Rata-rata:", rata_rata)
else:
    print("Tidak ada data yang dimasukkan.")
```

```python
def kata_unik(nama_file):
    try:
        with open(nama_file, "r") as file:
            unik = set()

            for baris in file:
                baris = baris.lower()
                kata_list = baris.split()

                for kata in kata_list:
                    kata = kata.strip(".,!?\"'()[]{}:;")
                    if kata:
                        unik.add(kata)

        print("\nKata unik dalam file:")
        for kata in sorted(unik):
            print(kata)

        print("\nTotal kata unik:", len(unik))

    except FileNotFoundError:
        print("File tidak ditemukan. Pastikan nama file benar.")
    except Exception as e:
        print("Terjadi error:", e)

nama_file = input("Masukkan nama file: ")
kata_unik(nama_file)
```
